```python
# @param {type:"string"}
start_url = "https://github.com/orcunmadran/miniOrumcek"

# @param {type:"slider", min:1, max:5, step:1}
depth_level = 3

print(f"Başlangıç URL'si: {start_url}")
print(f"Tarama Derinliği: {depth_level}")
```

```python
import requests
from bs4 import BeautifulSoup
import time

def web_spider(url, depth):
    visited = []  # Ziyaret edilen sayfalar
    to_visit = [(url, 0)]  # Ziyaret edilmesi gereken sayfalar (url, seviye)
    results = []  # Sonuçları kaydetmek için

    while to_visit:
        current_url, current_depth = to_visit.pop(0)

        if current_depth > depth:
            break

        if current_url not in visited:
            visited.append(current_url)
            try:
                response = requests.get(current_url)
                soup = BeautifulSoup(response.text, 'html.parser')
                title = soup.title.string if soup.title else 'No Title'

                # Ziyaret edilen sayfanın bilgilerini kaydet
                results.append(f'Seviye {current_depth}: {current_url} - Başlık: {title}')
                print(f'Ziyaret edilen: {current_url} - Başlık: {title}')

                if current_depth < depth:
                    for link in soup.find_all('a', href=True):
                        abs_link = requests.compat.urljoin(current_url, link['href'])
                        if abs_link not in visited:
                            to_visit.append((abs_link, current_depth + 1))
                time.sleep(1)  # Sunucuları yormamak için kısa bir bekleme

            except requests.exceptions.RequestException as e:
                print(f'Hata: {e}')

    # Sonuçları bir dosyaya yazma
    with open('crawl_results.txt', 'w') as file:
        for result in results:
            file.write(result + '\n')

    print("Tarama tamamlandı, sonuçlar 'crawl_results.txt' dosyasına kaydedildi.")
```

```python
web_spider(start_url, depth_level)
```
